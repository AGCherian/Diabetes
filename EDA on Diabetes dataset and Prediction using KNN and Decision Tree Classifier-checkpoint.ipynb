### Model -1

#### Prediction of Diabetes using KNN Model

1) Finding the hyperparameter k, the optimum number of neighbours

2) Fitting the best model on test data and finding the Accuracy, Recall and Precision

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initial Test run with default k=5
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_valid)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,f1_score, confusion_matrix, precision_score, recall_score, accuracy_score
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print("Confusion Matrix is:\n",confusion_matrix(y_valid,y_pred))

print("Precision Score is:\n",precision_score(y_valid,y_pred))
print("Recall Score is:\n",recall_score(y_valid,y_pred))
print("Accuracy Score is:\n",accuracy_score(y_valid,y_pred))
print("F1 Score is:\n",f1_score(y_valid,y_pred))

```

#### Finding Best K

```python
for i in range(1,11):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_valid)
    print("------------------For Neighbours= ",i,"------------------------")
    print("\nConfusion Matrix is:\n",confusion_matrix(y_valid,y_pred))

    print("\nPrecision Score is:\n",precision_score(y_valid,y_pred))
    print("\nRecall Score is:\n",recall_score(y_valid,y_pred))
    print("\nAccuracy Score is:\n",accuracy_score(y_valid,y_pred))
    print("\nF1 Score is:\n",f1_score(y_valid,y_pred))

```

### Thus it can be seen that F1 score is highest when No. of neighbours=5

### Model 2 - Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

#### Finding best depth

```python
for i in range(1,11):
    dtree = DecisionTreeClassifier(max_depth=i,criterion='gini')
    dtree.fit(X_train,y_train)
    y_pred=dtree.predict(X_valid)
    print("\n----------------------For Depth= ",i,"-------------------------\n")
    print("\nConfusion Matrix is:\n",confusion_matrix(y_valid,y_pred))

    print("\nPrecision Score is:\n",precision_score(y_valid,y_pred))
    print("\nRecall Score is:\n",recall_score(y_valid,y_pred))
    print("\nAccuracy Score is:\n",accuracy_score(y_valid,y_pred))
    print("\nF1 Score is:\n",f1_score(y_valid,y_pred))

    
```

### Thus it can be seen that F1 score is highest when No. of depth=8 ¶

#### Model -2 Decision Tree at depth 8 is performing better compared to Model -1 KNN at various values of k

Thus selecting Model-2 Decision Tree at depth 8 as final model, Lets see the results on the test data

```python
dtree = DecisionTreeClassifier(max_depth=8,criterion='gini')
dtree.fit(X_train,y_train)
y_pred=dtree.predict(X_test)
print("\n----------------------For Depth= ",8,"-------------------------\n")
print("\nConfusion Matrix is:\n",confusion_matrix(y_test,y_pred))

print("\nPrecision Score is:\n",precision_score(y_test,y_pred))
print("\nRecall Score is:\n",recall_score(y_test,y_pred))
print("\nAccuracy Score is:\n",accuracy_score(y_test,y_pred))
print("\nF1 Score is:\n",f1_score(y_test,y_pred))
```
